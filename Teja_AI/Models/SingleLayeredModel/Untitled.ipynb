# Dense Layer

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
# Create a one-layer model
model = Sequential()
model.add(Dense(64,activation = 'relu',input_shape = (20,)))
model.summary()
```

```python
# Compile the model (specify optimizer, loss function, etc.)
model.compile(optimizer='adam', loss='mse')

# Prepare the input data (replace this with your input data)
input_data = tf.random.normal((10, 20))  # A single input example with 10 features
```

```python

# Make predictions using the model
predictions = model.predict(input_data)
print("Predictions:", predictions)
predictions.shape
```

```python
import matplotlib.pyplot as plt
plt.plot(predictions)
plt.show()
```

# Conv2D 

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense

# Function to create and compile the model
def create_simple_conv2d_model(input_shape):
    model = Sequential()
    model.add(Conv2D(filters=16, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
    model.add(Flatten())
    model.add(Dense(1, activation='linear'))  # Assuming a single output prediction

    model.compile(optimizer='adam', loss='mse')  # Using Mean Squared Error (MSE) as the loss function

    return model

# Example input data
input_shape = (28, 28, 1)  # Adjust this based on your actual image dimensions (e.g., 28x28 pixels with one channel)

# Assuming you have a single input image
input_data = np.random.rand(1, *input_shape)

# Create the model
model = create_simple_conv2d_model(input_shape)

# Make a prediction using the model
prediction = model.predict(input_data)

print("Input data shape:", input_data.shape)
print("Output prediction:", prediction.shape)
```

# Conv3D

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv3D, Flatten, Dense

# Function to create and compile the model
def create_simple_conv3d_model(input_shape):
    model = Sequential()
    model.add(Conv3D(filters=16, kernel_size=(3, 3, 3), activation='relu', input_shape=input_shape))
    model.add(Flatten())
    model.add(Dense(1, activation='linear'))  # Assuming a single output prediction

    model.compile(optimizer='adam', loss='mse')  # Using Mean Squared Error (MSE) as the loss function

    return model

# Example input data
input_shape = (10, 32, 32, 1)  # Adjust the input shape based on your actual input volume dimensions
input_data = np.random.rand(1, *input_shape)

# Create the model
model = create_simple_conv3d_model(input_shape)

# Make a prediction using the model
prediction = model.predict(input_data)

print("Input data shape:", input_data.shape)
print("Output prediction:", prediction)
```

# Maxpool

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GlobalMaxPooling1D, GlobalAveragePooling1D

# Function to create and compile the model
def create_simple_model_with_pooling(input_length):
    model = Sequential()
    model.add(Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=(input_length, 1)))
    model.add(GlobalMaxPooling1D())  # Global Max Pooling

    model.compile(optimizer='adam', loss='mse')  # Using Mean Squared Error (MSE) as the loss function

    return model

# Example input data
input_length = 10  # Adjust this value based on your actual input length
input_data = np.random.rand(1, input_length, 1)  # Assuming a single input sequence

# Create the model
model = create_simple_model_with_pooling(input_length)

# Make a prediction using the model
prediction = model.predict(input_data)

print("Input data shape:", input_data.shape)
print("Output prediction:", prediction)
```

# Activation functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) 

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

# Generate random inputs
random_input = np.linspace(-5, 5, 100)

# Evaluate activation functions
sigmoid_output = sigmoid(random_input)
relu_output = relu(random_input)
tanh_output = tanh(random_input)
softmax_output = softmax(random_input)
leaky_relu_output = leaky_relu(random_input)

# Plot the graphs
plt.figure(figsize=(8, 6))

# Sigmoid
plt.subplot(2, 3, 1)
plt.plot(random_input, sigmoid_output)
plt.title('Sigmoid Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# ReLU
plt.subplot(2, 3, 2)
plt.plot(random_input, relu_output)
plt.title('ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Tanh
plt.subplot(2, 3, 3)
plt.plot(random_input, tanh_output)
plt.title('Tanh Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Softmax
plt.subplot(2, 3, 4)
plt.plot(random_input, softmax_output)
plt.title('Softmax Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Leaky ReLU
plt.subplot(2, 3, 5)
plt.plot(random_input, leaky_relu_output)
plt.title('Leaky ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.tight_layout()
plt.show()
```
